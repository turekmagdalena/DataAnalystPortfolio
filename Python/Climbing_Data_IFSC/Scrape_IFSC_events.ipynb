```python
import google_colab_selenium as gs

driver = gs.Chrome()
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import time
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
events_df = pd.DataFrame(columns=['Athlete', 'Athlete_Country', 'Rank', 'Event_Name', 'Discipline', 'Category'])


for i in range(1400, 1450):

    for disc in ['lead', 'boulder', 'speed', 'boulder&lead']:
      url = 'https://ifsc.results.info/#/event/' + str(i) + '/general/' + disc

      driver.get(url)

      try:
          # Wait for the event name element to be visible
          event_name_element = WebDriverWait(driver, 10).until(
              EC.visibility_of_element_located((By.CSS_SELECTOR, 'div.gr-event-name'))
          )

          # Get the event name text
          event_name = event_name_element.text
          try:
            elements = driver.find_elements(By.CSS_SELECTOR,'div.mx-2.nav-item' )

            if elements:
              print(str(i) + " " + event_name + " " + disc)
              print(f"Page {i} has {len(elements)} element(s) to click.")
              # Iterate over each element and click on it
              for element in elements:
                element.click()

                print(f"Clicked on element with text: {element.text}")

                time.sleep(2)

                # Find all athlete elements
                athlete_elements = driver.find_elements(By.CSS_SELECTOR, 'div.athlete.athlete-line.mobile')

                if athlete_elements:
                            print(f"Page {i} has {len(athlete_elements)} athlete(s).")

                            # Iterate over each athlete element
                            for athlete in athlete_elements:
                                # Extract rank
                                rank_script = "return arguments[0].querySelector('div.rank').innerText;"
                                rank = driver.execute_script(rank_script, athlete)

                                # Extract athlete name
                                name_script = "return arguments[0].querySelector('a.r-name').innerText;"
                                name = driver.execute_script(name_script, athlete)

                                # Extract country code
                                country_code_script = "return arguments[0].querySelector('span:last-child').innerText;"
                                country_code = driver.execute_script(country_code_script, athlete)


                                events_df.loc[len(events_df)] = [name,country_code, rank, event_name, disc, element.text ]
                                print(f"Rank: {rank}, Name: {name}, Country Code: {country_code}")
                else:
                            print(f"No athlete elements found on page {i}. Skipping...")

            else:
              print(f"No elements found on page {i} {disc}. Skipping...")

          except:
            continue




      except NoSuchElementException:
          print("Error: Element not found for " + str(i))
      except Exception as e:
          print("Error: " + str(i))
```
