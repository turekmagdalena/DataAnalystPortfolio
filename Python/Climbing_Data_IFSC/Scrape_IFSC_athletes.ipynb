```python
import google_colab_selenium as gs

driver = gs.Chrome()
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
```

```python
athletes_df = pd.DataFrame(columns=['Name', 'Country', 'Age', 'Height'])

for i in range(1, 17500):
  url = 'https://ifsc.results.info/#/athlete/' + str(i)

  # Open the URL
  driver.get(url)
  driver.implicitly_wait(10)

  try:

      #extract athlete name
      element = driver.find_element(By.CSS_SELECTOR, 'div.athlete-name')
      athlete_name = element.text.strip()

      #extract country code
      element = driver.find_element(By.CSS_SELECTOR, 'div.country-code')
      country_code = element.text.strip()

      element = driver.find_element(By.CSS_SELECTOR, 'div.athlete-info-container')
      info = element.text.strip()
      age = info[5:info.index('HEIGHT')-1] #extract age from athlete info
      height = info[info.index('HEIGHT')+7:info.index('ACTIVE')-1] #extract height from athlete info

      #append to athletes dataset
      athletes_df.loc[len(athletes_df)] = [athlete_name, country_code, age, height]

  except NoSuchElementException:
    print(f'Athlete number: {i}' ) #for troubleshooting purposes

  except Exception as e:
        print(f"Error: {e}")
        break
```
