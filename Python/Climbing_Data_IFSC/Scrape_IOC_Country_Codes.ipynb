```python
import google_colab_selenium as gs

driver = gs.Chrome()
```

```python
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import time
```

```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
```

```python
url = "https://simple.wikipedia.org/wiki/List_of_IOC_country_codes"

driver.get(url)
```

```python
codes = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][2]/tbody/tr/td[1]')
names = driver.find_elements(By.XPATH, '//table[@class="wikitable sortable jquery-tablesorter"][3]/tbody/tr/td[2]')
```

```python
ioc_codes = pd.DataFrame(columns=['Code', 'Name'])

for i in range(1,25):
  codes = driver.find_elements(By.XPATH,
                               f'//table[@class="wikitable sortable jquery-tablesorter"][{str(i)}]/tbody/tr/td[1]')
  names = driver.find_elements(By.XPATH,
                              f'//table[@class="wikitable sortable jquery-tablesorter"][{str(i)}]/tbody/tr/td[2]')

  for j,k in zip(codes,names):
    print(j.text, k.text)
    ioc_codes.loc[len(ioc_codes)] = [j.text, k.text]
```
